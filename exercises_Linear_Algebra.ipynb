```python
# Given a set of points in 2D, find the line of best fit using
# Numpy and plot the data points and the line using Matplotlib.
```

```python
x = np.random.randint(19, 41, size=25)
```

```python
x
```

```python
y = np.random.randint(27, 93, size=25)
```

```python
y
```

```python
m, c = np.polyfit(x, y, 1)
```

```python
plt.scatter(x, y)
plt.plot(x, m * x + c, color='g')
plt.show()
```

```python
# Given a set of points in 3D, find the plane of best fit using
# Numpy and plot the data points and the plane using Matplotlib.
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
z = np.random.randint(17, 63, size=25)
```

```python
z
```

```python
plane = np.column_stack((x, y, np.ones_like(x)))
```

```python
coeff, _, _, _ = np.linalg.lstsq(plane, z, rcond=None)
```

```python
xx, yy = np.meshgrid(x, y)
```

```python
zz = coeff[0] * xx + coeff[1] * yy + coeff[2]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)
ax.plot_surface(xx, yy, zz, alpha=0.5)
plt.show()
```
