```python
# Generate a 2D array of random integers between 1 and 100
# with a shape of (5, 10). Sort each row in ascending order and
# plot a heatmap of the sorted array.
```

```python
arr = np.random.randint(1, 101, size=(5, 10))
```

```python
arr
```

```python
arr_sort = np.sort(arr, axis=1)
```

```python
arr_sort
```

```python
plt.imshow(arr_sort, cmap='Blues')
plt.colorbar()
plt.show()
```

```python
# Generate a 2D array of random floats between 0 and 1
# with a shape of (100, 2). Sort the array by the first
# column in descending order and plot a scatter plot of the sorted values.
```

```python
arr = np.random.random((100, 2))
```

```python
arr
```

```python
index_sort = np.argsort(arr[:, 0])[::-1]
```

```python
arr_sort = arr[index_sort]
```

```python
arr_sort
```

```python
plt.scatter(arr_sort[:, 0], arr_sort[:, 1])
plt.show()
```

```python
# Sort the following array in ascending order and count the number of
# occurrences of each unique value. Plot a bar chart of the unique
# values and their counts.
```

```python
arr = np.random.randint(1, 16, size=50)
```

```python
arr
```

```python
unique_values, count = np.sort(np.unique(arr, return_counts=True))
```

```python
plt.bar(unique_values, count)
plt.show()
```
