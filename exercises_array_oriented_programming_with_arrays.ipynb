```python
# Create a contour plot of a Numpy 2D array, with custom levels
# and labels for the x-axis and y-axis.
```

```python
# Create a 3D surface plot of a Numpy 2D array,
# with a custom colormap and labels for the x-axis and y-axis.
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
arr = np.random.randint(1, 11, size=(3, 3))
```

```python
arr
```

```python
x_labels = ['Happy', 'Cute', 'Kind']
y_labels = ['a', 'b', 'c']
colormap = plt.cm.get_cmap('jet')
```

```python
x, y = np.meshgrid(np.arange(arr.shape[0]), np.arange(arr.shape[1]))
```

```python
x, y = x.flatten(), y.flatten()
```

```python
z = arr.flatten()
```

```python
fig = plt.figure()
surface = fig.add_subplot(111, projection='3d')
surface.plot_trisurf(x, y, z, cmap=colormap, edgecolor='none')
surface.set_xlabel('x - values')
surface.set_ylabel('y - values')
surface.set_zlabel('z - values')
plt.xticks(np.arange(len(x_labels)), x_labels)
plt.yticks(np.arange(len(y_labels)), y_labels)
plt.show()
```
